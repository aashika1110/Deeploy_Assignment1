```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images
def show(image, n, m, i, Title):
    plt.subplot(n, m, i)
    plt.imshow(image, cmap='gray')  # Keep cmap='gray'
    plt.title(Title)
    plt.axis('off')

# Load two images and resize them
image1 = cv2.imread('img1ass1.png', cv2.IMREAD_COLOR)  # Replace with first image path
image2 = cv2.imread('img2ass1.png', cv2.IMREAD_COLOR)  # Replace with second image path
image1 = cv2.resize(image1, (300, 300))
image2 = cv2.resize(image2, (300, 300))

# Convert images to grayscale for uniform visualization
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Apply a high-pass filter (Laplacian) to the first image
high_pass = cv2.Laplacian(gray1, cv2.CV_64F)
high_pass = np.uint8(np.absolute(high_pass))  # Convert to uint8

# Apply a low-pass filter (Gaussian Blur) to the second image
low_pass = cv2.GaussianBlur(gray2, (21, 21), 0)

# Combine the two filtered images to create the hybrid image
hybrid_image = cv2.addWeighted(high_pass, 0.5, low_pass, 0.5, 0)

# Save the final hybrid image to a file
output_path = 'output_hybrid_image.png'
cv2.imwrite(output_path, hybrid_image)  # Save the output image
print(f"Hybrid image saved to {output_path}")

# Plot the results
plt.figure(figsize=(10, 10))

show(gray1, 2, 3, 1, 'Original Image 1 (Grayscale)')
show(gray2, 2, 3, 2, 'Original Image 2 (Grayscale)')
show(high_pass, 2, 3, 3, 'High-Pass Filtered')
show(low_pass, 2, 3, 4, 'Low-Pass Filtered')
show(hybrid_image, 2, 3, 5, 'Hybrid Image')

plt.tight_layout()
plt.show()


```
